```python
import re
```

```python
# 2.1
pattern = "mailto.*?\>(\w+\d+)\<"
regex = re.compile(pattern)

filename = "pds_2012_roster.html"
file = open(filename, "r")

for line in file:
    result = regex.search(line)
    if result:
        print result.group(1)
```

```python
# 2.2
pattern = "bio-info.*?\>(.*?),(.*\n)+?.*?mailto.*?\>(\w+\d+)\<"
regex = re.compile(pattern)

filename = "pds_2012_roster.html"
file = open(filename, "r")
text = file.read()

allcount = 0
keepcount = 0

results = re.findall(pattern, text, re.MULTILINE)
for result in results:
    allcount = allcount + 1
    name = result[0]

    if len(name) <= 4:
        print result[2]
        keepcount = keepcount + 1
print "removed %d students" % (allcount - keepcount)
```

```python
# 2.3
pattern = "bio-info.*?\>(.*?), (.*?)\<(.*\n)+?.*?mailto.*?\>(\w+\d+)\<"
regex = re.compile(pattern)

filename = "pds_2012_roster.html"
file = open(filename, "r")
text = file.read()


results = re.findall(pattern, text, re.MULTILINE)
for result in results:
    print "%s\t%s\t%s" % (result[1], result[0], result[3])
```
