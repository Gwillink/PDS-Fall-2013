Analyzing NYC's Citi Bike Usage

Let's read the file into a python data stucture.

```python
file = open("citi_bike.csv", "r")
# read in the header line
headers = file.readline().strip().split(",")

days = [] # store a list of days we observe
data = [] # store the actual data as a list of lists
          # note that I dont think this is the best way to store things
    
for line in file:
    parts = line.strip().split(",")
    data.append(parts)

# take a peek at the first 5 entries
print data[0:5]
```

```python
# What's the distribution of 7 day pass purchases over time?
seven_day_purch = []
for d in data:
    seven_day_purch.append(d[-1])
print seven_day_purch
```

```python
# let's plot it!
import matplotlib.pyplot as plt
plt.plot(seven_day_purch, 'r')
```

```python
# What about 24 hour passes?
# What about annual memberships? 

annual_memberships = []
twentyfour_memberships = []

for d in data:
    twentyfour_memberships.append(d[-2])
    annual_memberships.append(d[-3])

plt.plot(seven_day_purch, 'r')
plt.plot(annual_memberships, 'g')
plt.plot(twentyfour_memberships, 'b')
```
