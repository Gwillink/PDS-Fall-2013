```python
# question b:
import json
filename = 'assess1_data.json'
file = open(filename, 'r')

namecounts = {}
for line in file:
    record = json.loads(line.strip())
    name = record['NAME']
    if name in namecounts:
        namecounts[name] = namecounts[name] + 1
    else:
        namecounts[name] = 1

for name,ct in namecounts.iteritems():
    if ct > 1:
        print "%s: %d" % (name, ct)
```

```python
# question d:
import json
filename = 'assess1_data.json'
file = open(filename, 'r')

RFA_2F = {}
RFA_2A = {}
WEALTH_INDEX = {}

for line in file:
    record = json.loads(line.strip())
    shouldfilter = False
    for k,v in record.iteritems():
        if str(v) == "-9999":
            shouldfilter = True
    if not shouldfilter:
        rfa2f = record['RFA_2F']
        rfa2a = record['RFA_2A']
        wealth = record['WEALTH_INDEX']
        
        if rfa2f in RFA_2F:
            RFA_2F[rfa2f] = RFA_2F[rfa2f] + 1
        else:
            RFA_2F[rfa2f] = 1

        if rfa2a in RFA_2A:
            RFA_2A[rfa2a] = RFA_2A[rfa2a] + 1
        else:
            RFA_2A[rfa2a] = 1
        
        if wealth in WEALTH_INDEX:
           WEALTH_INDEX[wealth] = WEALTH_INDEX[wealth] + 1
        else:
            WEALTH_INDEX[wealth] = 1
            
print "RFA_2F freqs: %s" % RFA_2F
print "RFA_2A freqs: %s" % RFA_2A

print "WEALTH_INDEX values: %s" % WEALTH_INDEX.keys()
```

```python
# question e:
import json
filename = 'assess1_data.json'
file = open(filename, 'r')

TARGET_B = {}
total = 0;

for line in file:
    record = json.loads(line.strip())
    shouldfilter = False
    for k,v in record.iteritems():
        if str(v) == "-9999":
            shouldfilter = True
    if not shouldfilter:
        target = record['TARGET_B']
        if target in TARGET_B:
            TARGET_B[target] = TARGET_B[target] + 1
        else:
            TARGET_B[target] = 1
        total = total + 1.0

for k,v in TARGET_B.iteritems():
    print "%s appears %d times, %f of the total" % (k, v, v/total)
```

```python
# question f:
# compute the mean (for instance) for each responder group.

import json
filename = 'assess1_data.json'
file = open(filename, 'r')

target_counts = {}
target_wealth = {}
total = 0;

for line in file:
    record = json.loads(line.strip())
    shouldfilter = False
    for k,v in record.iteritems():
        if str(v) == "-9999":
            shouldfilter = True
    if not shouldfilter:
        target = record['TARGET_B']
        wealth = record['WEALTH_INDEX']
        if target in target_counts:
            target_counts[target] = target_counts[target] + 1
            target_wealth[target] = target_wealth[target] + wealth
        else:
            target_counts[target] = 1
            target_wealth[target] = wealth
        total = total + 1.0

for k in target_counts.keys():
    print "%s mean wealth: %f" % (k, target_wealth[k]/target_counts[k])
```

```python
# question g:
%matplotlib inline
import json
import matplotlib.pyplot as plt
filename = 'assess1_data.json'
file = open(filename, 'r')

zero_vals = []
one_vals = []

for line in file:
    record = json.loads(line.strip())
    shouldfilter = False
    for k,v in record.iteritems():
        if str(v) == "-9999":
            shouldfilter = True
    if not shouldfilter:
        target = record['TARGET_B']
        wealth = record['WEALTH_INDEX']
        if target == 1:
            one_vals.append(wealth)
        else:
            zero_vals.append(wealth)

plt.hist([zero_vals, one_vals], normed=True, bins = 20, stacked=False, color = ['r', 'b'])
plt.legend(['non responders', 'responders'])
plt.show()
```

```python
# question i:
import json
filename = 'assess1_data.json'
file = open(filename, 'r')

records = {}

for line in file:
    record = json.loads(line.strip())
    shouldfilter = False
    for k,v in record.iteritems():
        if str(v) == "-9999":
            shouldfilter = True
    if not shouldfilter:
        name_parts = record['NAME'].split(' ')
        name = name_parts[-1]
        # remove any suffixes
        if name == 'Sr.' or name == 'Jr.' or name == 'III':
            name = name_parts[-2]
        records[line.strip()] = name
 
# found this on google:
output = sorted(records.items(), key=lambda x: x[1])
print "first 10"
for rec in output[0:10]:
    print rec[0]
print "20000 - 20010"
for rec in output[20000:20010]:
    print rec[0]
```
