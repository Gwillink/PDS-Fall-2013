```python
# clean up some badly parsed text
expmt_results = expmt_results_raw.replace('\noff', 'off')
expmt_results = expmt_results.replace('\non', 'on')
expmt_results.head()
```

```python
# remove some columns i don't want
del expmt_results['pages_per_visit']
del expmt_results['visit_shop_fav']
del expmt_results['visit_error_page']
del expmt_results['visit_search']
del expmt_results['mean_searches']
```

```python
# clean up a column named badly
expmt_results = expmt_results.rename(columns={'mean_listing_views': 'total_listing_views'})
expmt_results
```

```python
# add a couple rates
expmt_results_rates = expmt_results.copy()
expmt_results_rates ['conv_rate'] = expmt_results_rates ['conversions'] / expmt_results_rates ['visit_count'].astype(float64)
expmt_results_rates ['cart_view_exit_rate'] = expmt_results_rates ['cart_view_exits'] / expmt_results_rates ['visit_count'].astype(float64)
expmt_results_rates ['mean_listing_views'] = expmt_results_rates ['total_listing_views'] / expmt_results_rates ['visit_count'].astype(float64)
```

### Move into R for more visual inspection, via ggplot

```python
# Make a helper function to Takes in pandas dataframe and name you want to give it for R
# Converts to R object and pushes into R environment
def add_pandas_df_to_r(pd_df, r_name):
    df_r = com.convert_to_r_dataframe(pd_df)
    r.assign(r_name, df_r)
```

```python
# melt the dataframe to make it friendly to ggplot
# push the melted dataframe to the R environment
expmt_results_rates_melt = pd.melt(expmt_results_rates, id_vars=['date','variant'])
add_pandas_df_to_r(expmt_results_rates_melt, 'expmt_results_r')
```

```python
%%R

library(scales)
library(ggplot2)

expmt_results_r$date <- as.Date(as.character(expmt_results_r$date))
expmt_results_r$variant <- as.factor(expmt_results_r$variant)
expmt_results_r$variable <- as.factor(expmt_results_r$variable)
expmt_results_r$value <- as.numeric(expmt_results_r$value)

rate <- subset(expmt_results_r, variable == 'conv_rate')

p <- ggplot(rate, aes(x = date, y = value, group = variant)) + 
    scale_colour_brewer(palette="Set1") + 
    geom_line(aes(color = variant), size = 0.5) + 
    scale_y_continuous("", limits=c(0.025, .032), labels = percent) + 
    scale_x_date("") + 
    ggtitle("Conversion Rate") + 
    theme_bw() +
    theme(legend.position = "right")
print(p)

mean <- subset(expmt_results_r, variable == 'mean_listing_views')

p2 <- ggplot(mean, aes(x = date, y = value, group = variant)) + 
    scale_colour_brewer(palette="Set1") + 
    geom_line(aes(color = variant), size = 0.5) + 
    scale_y_continuous("", limits=c(4, 5.5), labels = comma) + 
    scale_x_date("") + 
    ggtitle("Mean Listing Views") + 
    theme_bw() +
    theme(legend.position = "right")
print(p2)
# ggsave('/home/nthomas/files/overview_yylast18.pdf', scale=1.25)
```

### Check the Bucketing

```python
# make a dataframe with just visits by variant by day
bucketing_df = expmt_results[['variant','date','visit_count']]
bucketing_df
```

```python
# find daily totals, rename columns
date_sum = bucketing_df.groupby(['date']).sum()
date_sum = date_sum.rename(columns={'visit_count': 'total_visits'})
date_sum
```

```python
# merge together
bucketing_df = bucketing_df.merge(date_sum, left_on='date', right_index=True, how='left', sort=False)
bucketing_df
```

```python
bucketing_df['percent_visits'] = bucketing_df.visit_count / bucketing_df.total_visits.astype(float64)
bucketing_df
```

```python
bucketing_df_melt = pd.melt(bucketing_df, id_vars=['date','variant'])
add_pandas_df_to_r(bucketing_df_melt, 'bucketing_df_r')
```

```python
%%R
visit_count <- subset(bucketing_df_r, variable == 'visit_count')
perc_visits <- subset(bucketing_df_r, variable == 'percent_visits')
```

```python
%%R

library(scales)
library(ggplot2)

visit_count$date <- as.Date(as.character(visit_count$date))
visit_count$variant <- as.factor(visit_count$variant)
visit_count$variable <- as.factor(visit_count$variable)
visit_count$value <- as.numeric(visit_count$value)

p <- ggplot(visit_count, aes(x = date, y = value, group = variant)) + 
    scale_colour_brewer(palette="Set1") + 
    geom_line(aes(color = variant), size = 0.5) + 
    scale_y_continuous("", labels = comma) + 
    scale_x_date("") + 
    ggtitle("Visit Count") + 
    theme_bw() +
    theme(legend.position = "right")
print(p)

perc_visits$date <- as.Date(as.character(perc_visits$date))
perc_visits$variant <- as.factor(perc_visits$variant)
perc_visits$variable <- as.factor(perc_visits$variable)
perc_visits$value <- as.numeric(perc_visits$value)

p2 <- ggplot(perc_visits, aes(x = date, y = value, group = variant)) + 
    scale_colour_brewer(palette="Set1") + 
    geom_line(aes(color = variant), size = 0.5) + 
    scale_y_continuous("", labels = comma) + 
    scale_x_date("") + 
    ggtitle("Percent Visits") + 
    theme_bw() +
    theme(legend.position = "right")
print(p2)
```
